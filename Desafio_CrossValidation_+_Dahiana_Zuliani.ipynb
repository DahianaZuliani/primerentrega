```python
def dropearcolumnas(X):
  X=X.drop(['DATAFLOW','SEXO', 'edad', 'unit','geo','COMERCIAL'],axis=1)
  X.columns=['mes', 'femmas','años','cantidad', 'lugar', 'carrera']
  return X
```

```python
dropearcolumnas(X_train)
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
DropColumns=FunctionTransformer(dropearcolumnas)
```

```python
DropColumns.fit_transform(X_train)
```
