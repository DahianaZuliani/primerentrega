```python
x= df [['edad']]
y=df.OBS_VALUE
```

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
modelo=LR()
```

```python
modelo.fit(x,y)
```

```python
modelo.coef_
```

```python
modelo.intercept_
```

```python
m_sk=modelo.coef_[0]
b_sk=modelo.intercept_
```

```python
y_sk=m_sk*x+ b_sk
```

```python
plt.figure(figsize=[4,4])
plt.plot(x,df.edad,'ok')
plt.plot(x,y_sk,'r')
```
