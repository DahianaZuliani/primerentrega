```python
green_diamond= dict(markerfacecolor='g', marker= 'D')
fig, ax = plt.subplots(figsize=(10,5))
ax.boxplot(df['OBS_VALUE'],vert=False,notch=True,flierprops=green_diamond)
ax.set_title('Distribucion de salarios')
ax.set_xlabel('Salario')
```

Se puede observar que hay bastantes valores por bajo del salario minimo, ya que hay muchas personas que no pueden obtener un salario por encima de la media por distintas circustancias socioeconomicas


```python
ax = sns.regplot(x="TIME_PERIOD", y="OBS_VALUE", data=df)
```

```python
ax.set(xlabel='OBS_VALUE', ylabel='TIME_PERIOD')
plt.title('AÑO VS SALARIO')
```

OBTENEMOS EL CONTEO DEL PROBLEMA COMERCIAL DE QUE LOS SALARIOS SEAN POR DEBAJO DE LA MEDIA

```python

df_problema= df[df['OBS_VALUE']>=600].groupby('OBS_VALUE').size().reset_index(name='counts')
df_problema.sort_values(ascending=False,inplace=True,by='counts')
df_problema.head()
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.set_title('Top 10 problemas economicos')
sns.barplot(data=df_problema.head(10), y='OBS_VALUE',x='counts', orient='h')
ax.set_ylabel('OBS_VALUE')
ax.set_xlabel('numero de personas')
```
